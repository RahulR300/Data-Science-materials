```python
print(sns.__version__)
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.tail()
```

```python
#pairplot
sns.pairplot(df,hue='sex')
```

```python
df1=df.drop(['sex','smoker','day','time'],axis=1)
```

```python
sns.heatmap(df1.corr(),annot=True)
```

```python
#jointplot
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.displot(data=df,x='tip')
```

```python
sns.distplot(df['tip'],kde=False)
```

```python
#countplot
sns.countplot(x='sex', data=df)
```

```python
sns.countplot(y='sex', data=df)
```

```python
#barplot
sns.barplot(y='total_bill',x='sex',data=df)
```

```python
#boxplot
sns.boxplot(x='total_bill',y='sex',data=df,color='g')
```

```python
#violinplot
sns.violinplot(x='total_bill',y='day',data=df,palatte='rainbow')
```
