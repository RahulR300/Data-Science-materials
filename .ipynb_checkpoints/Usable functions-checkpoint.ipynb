# Function for pairwise correlation 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Supress Warnings
import warnings 
warnings.filterwarnings('ignore')


# Load your dataset (replace 'your_dataset.csv' with the actual filename)
df = pd.read_csv('train.csv')

# Compute pairwise correlations
correlation_matrix = df.corr()

# Plot heatmap of correlations
plt.figure(figsize=(10, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Pairwise Correlation Heatmap')
plt.show()

# Check for multicollinearity (optional)
threshold = 0.7  # You can adjust this threshold as needed
multicollinear_features = set()
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            colname_i = correlation_matrix.columns[i]
            colname_j = correlation_matrix.columns[j]
            multicollinear_features.add(colname_i)
            multicollinear_features.add(colname_j)

if multicollinear_features:
    print("Features with high multicollinearity:", multicollinear_features)
else:
    print("No features have high multicollinearity.")
```

# Seperating the Numerical & Categorical Variables

```python
import pandas as pd

def separate_features_by_dtype(df):
    categorical_features = df.select_dtypes(include=['object', 'category']).columns.tolist()
    numerical_features = df.select_dtypes(include=['int', 'float']).columns.tolist()
    return categorical_features, numerical_features

# Assuming 'df' is your DataFrame containing the dataset

# Load your dataset into 'data'
data = pd.read_csv("train.csv")

# Separate categorical and numerical features
categorical_features, numerical_features = separate_features_by_dtype(data)

print("Categorical features:")
print(categorical_features)
print("\n\nNumerical features:")
print(numerical_features)
```

```python
import pandas as pd

def separate_features(df):
    # Separate numerical and categorical features
    numerical_features = df.select_dtypes(include=['number']).columns.tolist()
    categorical_features = df.select_dtypes(exclude=['number']).columns.tolist()
    
    # Create new DataFrames for numerical and categorical features
    df_num = df[numerical_features]
    df_cat = df[categorical_features]
    
    return df_num, df_cat

# Example usage:
# Assuming 'df' is your DataFrame
# Replace 'df' with your actual DataFrame name

# Example DataFrame
df = pd.read_csv("train.csv")

# Separate features
df_num, df_cat = separate_features(df)
```

# Finding Missing values of every column

```python
import pandas as pd

def print_missing_values(df):
    # Calculate the count of missing values in each column
    missing_values_count = df.isnull().sum()
    
    # Print the features with missing values
    print("Features with Missing Values:")
    for feature, count in missing_values_count.items():
        if count > 0:
            print(f"{feature}: {count}")

# Example usage:
# Assuming 'df' is your DataFrame
# Replace 'df' with your actual DataFrame name

# Example DataFrame
df = pd.read_csv("train.csv")

# Print features with missing values
print_missing_values(df)
```

```python
# % of missing values in num_df

import pandas as pd

def print_missing_values_percentage(df):
    # Calculate the percentage of missing values in each column
    missing_values_percentage = (df.isnull().sum() / len(df)) * 100
    
    # Print the columns and their missing values percentage
    print("Percentage of Missing Values in Each Feature:")
    for column, percentage in missing_values_percentage.items():
        print(f"{column}: {percentage:.2f}%")

# Example usage:
# Assuming 'df' is your DataFrame
# Replace 'df' with your actual DataFrame name

# Example DataFrame
data = pd.DataFrame(df_num)

# Print percentage of missing values in each feature
print_missing_values_percentage(df)
```

```python
def count_variable_types(df):
    variable_types = df.dtypes
    numerical_count = sum(variable_types == 'int64') + sum(variable_types == 'float64')
    categorical_count = sum(variable_types == 'object') 
    
    return numerical_count, categorical_count

numerical, categorical = count_variable_types(df)
print("Number of numerical variables:", numerical)
print("Number of categorical variables:", categorical)
```

# Ploting Distributions of all the columns in a dataframe

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_distributions(dataframe):
    # Iterate through each column in the DataFrame
    for column in dataframe.columns:
        # Check if the column contains numerical data
        if dataframe[column].dtype in ['int64', 'float64']:
            # Create a distribution plot using Seaborn
            sns.distplot(dataframe[column].dropna(), kde=False)  # Drop NaN values for plotting
            plt.title(f'Distribution of {column}')
            plt.xlabel(column)
            plt.ylabel('Frequency')
            plt.show()

plot_distributions(dataframe)
```
