```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Function to detect and print outliers
def detect_and_print_outliers(df_num):
    for col in df_num.columns:
        if df[col].dtype in ['int64', 'float64']:  # Process only numerical columns
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            outliers = df[(df_num[col] < lower_bound) | (df_num[col] > upper_bound)]
            
            # Print the outliers
            print(f"Outliers in {col} ({len(outliers)} rows):")
            
            # Plot the data and highlight the outliers
            plt.figure(figsize=(10, 5))
            sns.boxplot(x=df[col], color='blue')
            sns.stripplot(x=outliers[col], color='red', size=8)
            plt.title(f"Boxplot with Outliers Highlighted for {col}")
            plt.xlabel(col)
            plt.show()


# Run the function on the data frame
detect_and_print_outliers(df_num)
```

```python
import pandas as pd

def median_imputation(df_num):
    # Identify numerical columns with missing values
    num_cols_with_missing = df_num.select_dtypes(include=['number']).columns[df_num.isnull().sum() > 0]
    
    # Compute the median of these columns
    for col in num_cols_with_missing:
        median_value = df[col].median()
        print(f"Imputing missing values in column '{col}' with median value {median_value}")
        
        # Replace missing values with the median
        df_num[col].fillna(median_value, inplace=True)
    
    return df_num


# Example DataFrame

df_imp= pd.DataFrame(df_num)

# Perform median imputation
df_imp = median_imputation(df_num)

# Display the DataFrame after median imputation
print("\nDataFrame after Median Imputation:")
print(df_imp)
```

```python
import pandas as pd

def mode_imputation(df_cat):
    # Identify categorical columns with missing values
    cat_cols_with_missing = df_cat.select_dtypes(include=['object', 'category']).columns[df_cat.isnull().sum() > 0]
    
    # Compute the mode of these columns
    for col in cat_cols_with_missing:
        mode_value = df[col].mode()[0]
        print(f"Imputing missing values in column '{col}' with mode value '{mode_value}'")
        
        # Replace missing values with the mode
        df_cat[col].fillna(mode_value, inplace=True)
    
    return df_cat

# Example DataFrame
df_mimp = pd.DataFrame(df_cat)

# Perform mode imputation on categorical features
df_mimp = mode_imputation(df_cat)

# Display the DataFrame after mode imputation
print("\nDataFrame after Mode Imputation:")
print(df_mimp)
```
