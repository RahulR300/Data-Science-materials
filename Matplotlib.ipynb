# Outliers in statistics

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
##Define our Datasets
datasets=[11,10,12,14,12,15,14,13,15,102,12,14,17,19,107,10,13,12,14,12,108,12,11,14,13,15,10,15,12,10,14,13,15,10]
```

```python
outliers=[]
def detect_outliers(data):
    threshold=3
    mean=np.mean(data)
    std=np.std(data)
    
    for i in data:
        z_score=(i-mean)/std
        if np.abs(z_score) > threshold:
            outliers.append(i)
        if z_score > threshold:
            print(z_score)
    return outliers
```

```python
detect_outliers(datasets)
```

```python
dataset=sorted(datasets)
dataset
```

```python
q1,q3=np.percentile(dataset,[25,75])
```

```python
print(q1,q3)
```

```python
iqr=q3-q1
iqr
```

```python
lower_fence = q1-(1.5*iqr)
upper_fence = q3-(1.5*iqr)
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import math
np.random.seed(6)
school_ages=stats.poisson.rvs(loc=18,mu=35,size=1500)
classA_ages=stats.poisson.rvs(loc=18,mu=30,size=60)
```

```python
school_ages
```

```python
classA_ages
```

```python
 classA_ages.mean()
```

```python
ttest_1samp(classA_ages,popmean=school_ages.mean())
```

```python
_,p_value=ttest_1samp(classA_ages,popmean=school_ages.mean())
```

```python
if p_value<0.05:
    print("Accept Ho")
else:
    print("Reject Ho")
```
