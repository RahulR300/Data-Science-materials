# Pandas Combining DataFrames

```python
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
tips=sns.load_dataset('Tips')
tips.head()
```

```python
#Merge - used for complex column wise combinations of df's same like sql way!! 
```

```python
tips_bill=tips.groupby(['sex','smoker'])[['total_bill','tip']].sum()
tips_tip=tips.groupby(['sex','smoker'])[['total_bill','tip']].sum()

del tips_bill['tip']
del tips_tip['total_bill']
```

```python
tips_bill
```

```python
tips_tip
```

```python
pd.merge(tips_bill, tips_tip,right_index=True,left_index=True)     #self validation if index is not given!
```

```python
pd.merge(
    tips_bill.reset_index(),
    tips_tip.reset_index(),
    on=['sex', 'smoker']
)
```

```python
pd.merge(
    tips_bill.reset_index(),
    tips_tip.reset_index())
```

```python
# it can merge on partial column and index
pd.merge(
    tips_bill.reset_index(), 
    tips_tip,
    left_on=['sex', 'smoker'],
    right_index=True
)
```

```python
#it can do interesting combinations
tips_bill_strange=tips_bill.reset_index(level=0)
tips_bill_strange
```

```python
pd.merge(
    tips_tip.reset_index(), 
    tips_bill_strange,
    on=['sex', 'smoker']
)
```

```python
pd.merge(
    tips_bill.reset_index(),
    tips_tip.reset_index().head(2),
    how='left')
```

```python
pd.merge(
        tips_bill.reset_index(),
        tips_tip.reset_index().head(2),
        how='inner')
```

```python
pd.merge(
        tips_bill.reset_index().tail(3),
        tips_tip.reset_index().head(2),
        how='outer',
        indicator=True)      #indicator
```

```python
pd.merge(tips_bill,
        tips_bill,
        right_index=True,            #same
        left_index=True,             ##name
        suffixes=('_left','_right'))
```

# Concatenation

```python
#It is used for combining of 2/more df's in either cols/rows wise


pd.concat([tips_bill,tips_bill,tips_tip],sort=False)  #row wise concat
```

```python
pd.concat([tips_bill,tips_tip],axis=1)   #col wise
```

```python
#add on dataset where it's from
pd.concat([tips_bill,tips_tip], sort=False, keys=['a','b'])  
```
